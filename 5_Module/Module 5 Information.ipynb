```python
# Create a smaller dataframe of just James Harden Gamelogs
harden = df[df["Name"] == "James Harden"]
harden = harden.reset_index(drop=True)
harden.head()
```

```python
# This plot shows how many points he scored in each game
# It looks super congested! 
plt = harden.plot(x="Date", y="PTS", kind="bar")
```

```python
# Let's look at a specific range of dates (first two weeks of december)
# Date format in the table is YYYY-MM-DD
dates = []
for day in range(14):
    date = "2018-12-" + str(day)
    dates.append(date)
print(dates)
```

```python
# Lets fix the code in the previous cell
dates = []
for day in range(14):
    if day < 10:
       str_day = "0" + str(day + 1)
    else:
        str_day = str(day + 1)
    date = "2018-12-" + str_day
    dates.append(date)
print(dates)
```

```python
# Lets replot our histogram
small_harden = harden[harden["Date"].isin(dates)]
small_harden
```

```python
plt = small_harden.plot(x="Date", y="PTS", kind="bar")
```
