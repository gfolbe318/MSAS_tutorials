```python
# Change the column_names!!!
list_cols = column_names.to_list()
list_cols[2] = "Team"
df.columns = list_cols
df.head()
```

```python
# Change the column names the proper way!
df.rename(columns = {
    "Tm" : "Team",
    "G" : "Games",
    "Sk" : "Sacks"
}, inplace = True)

df.head()
```

```python
df = df.drop(columns=["Rk"])
df.head()
```

```python
# Add a new column
df["New Col"] = df["Cmp"]
df.head()
```

```python
# Reset the index column
df = df.set_index("Player")
df.head()
```

```python
df.loc[["Matthew Stafford"]]
```

```python
qbs = df[df["Pos"] == "QB"]
qbs.head()
```

```python
qbs.shape
```

```python
all_qbs = df[(df["Pos"] == "QB") | (df["Pos"] == "qb")]
all_qbs.head()
```

```python
all_qbs.shape
```

```python
starters = df[(df["Pos"] == "QB") & (df["GS"] == 16)]
starters
```

```python
# Filter on list condition
NFC_South = ["NOR", "ATL", "TAM", "CAR"]
nfc_south_qbs = all_qbs[all_qbs["Team"].isin(NFC_South)]
nfc_south_qbs
```

```python
# Sort the values least to greatest
eligible = all_qbs[all_qbs["Games"] > 6]
eligible = eligible.sort_values("Cmp%", ascending=False)
eligible.head()
```
